## Regex

[мп]{1}а[мп]{1}[ауе]{1}  
[м|п]а[м|п][а|у|е]

\b[оО]ко\b  
(?i)\boко\b  
*выбирает "око" из строки "Око - это глаз", но не из "Молоко"*

а+ - символ повторяется от 1 до бесконечности    
а* - символ повторяется от 0 до бесконечности  
а? - символ встречается 0 или 1 раз

_сделать выбор_:  
[лей] - не учитывается порядок символов  
(?:ли|лей) - выберется либо "ли", либо "лей" + скобки не считаются захватом строки  
*Пример:*  руб(?:\.|лей)? ловит "руб", "руб.", "рублей"

\S{,2}ap.[th].*  
\S{,2}ap.[th].{,6}

[rb]?af[gfkh]{2}\s?[fka]\w{,4}  
[rb]?af\w{2} ?[fka]\w*  
[rb]?af\w{2} ?[fka]\w{,5}
